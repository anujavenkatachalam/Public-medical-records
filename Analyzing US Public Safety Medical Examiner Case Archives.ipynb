#### Title: Importing, cleaning and analyzing medical examiner records from Cook County, Chicago from https://datacatalog.cookcountyil.gov/Public-Safety/Medical-Examiner-Case-Archive/cjeq-bs86

```python
# Importing the necessary packages
import pandas as pd
import numpy as np
%matplotlib inline
import re
```

```python
# Importing and cleaning the dataset
df=pd.read_csv("Medical_Examiner_Case_Archive.csv", parse_dates=True, na_values=["-1"])
```

```python
# Cleaning the primary cause variable
df["Primary Cause"].value_counts(dropna=False).head(10)
```

```python
df=df.rename(columns={"Primary Cause":"Primary_cause"})
```

```python
df["missing_cause"]=df["Primary_cause"].isna()==True
df["missing_cause"].value_counts()

#Ans=124 are missing
```

```python
df["Primary_cause"].dropna(inplace=True)
```

```python
# Examining cardiovascular disease-related deaths
df["Cardio_deaths"]=df["Primary_cause"].str.contains("(CARDIO)", regex=True)
df[df["Cardio_deaths"]==True]
```

```python
df[df["Cardio_deaths"]==True].Primary_cause.value_counts()
```

```python
# Cleaning the variable to achieve consistency in coding values
df.Primary_cause.replace(to_replace=r".*(CARDIO).*", value="CARDIOVASCULAR DISEASE", regex=True, inplace=True)
df.Primary_cause
```

```python
# Re-checking the distribution of deaths by cause after having re-coded the cardiovascular disease-related deaths
df.Primary_cause.value_counts(normalize=True)*100
```

```python
# Finding deaths caused by fentanyl

df.Primary_cause.dropna(inplace=True)
df["fentanyl_deaths"]=df["Primary_cause"].str.contains(".*(fentanyl).*", regex=True, flags=re.IGNORECASE)
df[df["fentanyl_deaths"]==True]
```

```python
# Breaking it down by race

#First creating a new variable called other_causes which takes on a value of True if death is not caused by fentanyl.
df["other_causes"]=df["Primary_cause"].str.contains(".*(fentanyl).*")==False
```

```python
#Then, generating two queries which show deaths from fentanyl deaths and other causes by race. 
series=[
df[(df["fentanyl_deaths"]==True)].Race.value_counts(normalize=True)*100,
df[df["other_causes"]==True].Race.value_counts(normalize=True)*100,
df[df["Manner_of_Death"]=="ACCIDENT"].Race.value_counts(normalize=True)*100,
df[df["Manner_of_Death"]=="HOMICIDE"].Race.value_counts(normalize=True)*100,
df[df["Manner_of_Death"]=="SUICIDE"].Race.value_counts(normalize=True)*100
]
```

```python
#Then, creating a separate dataframe with these two queries
df_deaths_by_race=pd.concat(series,axis=1, sort=False)
df_deaths_by_race.reset_index(inplace=True)

df_deaths_by_race.columns = ["Race","Fetanyl_deaths","Other_causes","Accident","Homicide","Suicide"]
df_deaths_by_race
```

```python
# Plotting opioid deaths by year
df=df.rename(columns={
          "Case Number":"Case_Number"})

df[(df["Opioid Related"]==True) & (df["year"]!=2018)].groupby(by="year").Case_Number.count().sort_index(ascending=False).plot(x="year", kind="barh")
```

```python
# Plotting fentanyl deaths by year
df[(df["fentanyl_deaths"]==True) & (df["year"]!=2018)].groupby(by="year").Case_Number.count().sort_index(ascending=False).plot(x="year", kind="barh")
```
